**K-MEANS CLUSTERING**

```r
install.packages("ggrepel")
```

```r
library(ggplot2)
library(ggrepel)

data <- read.csv("Fishing.csv")

data_numeric <- data[sapply(data, is.numeric)]

data_scaled <- scale(data_numeric)

set.seed(123)
k <- 3
kmeans_result <- kmeans(data_scaled, centers = k, nstart = 25)

data$cluster <- as.factor(kmeans_result$cluster)
centroids <- as.data.frame(kmeans_result$centers)
colnames(centroids) <- colnames(data_numeric)
centroids$cluster <- as.factor(1:k)

ggplot(data, aes(x = data_numeric[,1], y = data_numeric[,2], color = cluster)) +
  geom_point(size = 3, alpha = 0.7) +
  geom_point(data = centroids, aes_string(x = colnames(data_numeric)[1], y = colnames(data_numeric)[2]),color = "red", size = 5, shape = 4) +
  labs(title = "21BDS0364 - K-means Clustering with Centroids", x = colnames(data_numeric)[1], y = colnames(data_numeric)[2]) +
  theme_minimal()
```

**HIERARCHICAL CLUSTERING**

```r
install.packages("dendextend")
```

```r
library(ggplot2)
library(dendextend)

data <- read.csv("Fishing.csv")

data_numeric <- data[sapply(data, is.numeric)]

data_scaled <- scale(data_numeric)

distance_measures <- c("euclidean", "manhattan", "maximum", "canberra", "minkowski", "binary")

par(mfrow = c(2, 3))
for (dist_method in distance_measures) {
  dist_matrix <- dist(data_scaled, method = dist_method)

  hc <- hclust(dist_matrix, method = "ward.D")

  plot(as.dendrogram(hc), main = paste("21BDS0364 - Dendrogram using", dist_method, "distance"),
  xlab = "", sub = "", ylab = "Height")
}
```

**PRINCIPLE COMPONENT ANALYSIS**

```r
library(ggplot2)
data <- read.csv("Fishing.csv")
data_numeric <- data[sapply(data, is.numeric)]
data_scaled <- scale(data_numeric)
pca_result <- prcomp(data_scaled, center = TRUE, scale. = TRUE)
summary(pca_result)
```

```r
print(pca_result$rotation)
```

```r
scree_data <- data.frame(
  Component = paste0("PC", 1:length(pca_result$sdev)),
  Variance = (pca_result$sdev)^2 / sum((pca_result$sdev)^2) * 100
)

ggplot(scree_data, aes(x = Component, y = Variance)) +
  geom_bar(stat = "identity", fill = "skyblue") +
  geom_line(aes(group = 1), color = "red") +
  geom_point(color = "red") +
  labs(title = "21BDS0364 - Scree Plot", x = "Principal Component", y = "Variance Explained (%)") +
  theme_minimal()
```

```r
pca_data <- data.frame(PC1 = pca_result$x[, 1], PC2 = pca_result$x[, 2])

ggplot(pca_data, aes(x = PC1, y = PC2)) +
  geom_point(size = 3, alpha = 0.7, color = "blue") +
  labs(title = "21BDS0364 - PCA, First Two Principal Components", x = "PC1", y = "PC2") +
  theme_minimal()
```
