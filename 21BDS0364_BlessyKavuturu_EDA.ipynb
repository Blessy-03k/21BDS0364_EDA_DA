```r
library(ggplot2)
library(dplyr)
```

```r
data <- read.csv("Fishing.csv")
head(data, 20)
```

```r
dim(data)
```

```r
str(data)
```

```r
summary(data)
```

```r
sum(is.na(data))
```

```r
sum(duplicated(data))
```

```r
ggplot(data, aes(x = price)) +
  geom_histogram(binwidth = 10, fill = "steelblue", color = "black", alpha = 0.7) +
  labs(title = "Price Distribution", x = "Price", y = "Frequency") +
  theme_minimal()
```

```r
ggplot(data, aes(x = catch)) +
  geom_histogram(binwidth = 0.1, fill = "steelblue", color = "black", alpha = 0.7) +
  labs(title = "Catch Distribution", x = "Catch", y = "Frequency") +
  theme_minimal()
```

```r
ggplot(data, aes(x = income)) +
  geom_histogram(binwidth = 500, fill = "steelblue", color = "black", alpha = 0.7) +
  labs(title = "Income Distribution", x = "Income", y = "Frequency") +
  theme_minimal()
```

```r
ggplot(data, aes(x = mode, y = price, fill = mode)) +
  geom_boxplot() +
  labs(title = "Price by Fishing Mode", x = "Fishing Mode", y = "Price") +
  theme_minimal()
```

```r
ggplot(data, aes(x = mode, y = catch, fill = mode)) +
  geom_boxplot() +
  labs(title = "Catch by Fishing Mode", x = "Fishing Mode", y = "Catch") +
  theme_minimal()
```

```r
ggplot(data, aes(x = price, y = catch)) +
  geom_point(color = "blue", alpha = 0.5) +
  labs(title = "Price vs Catch", x = "Price", y = "Catch") +
  theme_minimal()
```

```r
ggplot(data, aes(x = price, y = catch, color = mode)) +
  geom_point(alpha = 0.7) +
  labs(title = "Price vs Catch by Fishing Mode", x = "Price", y = "Catch") +
  theme_minimal()
```

```r
ggplot(data, aes(x = price, y = income, color = mode)) +
  geom_point(alpha = 0.7) +
  labs(title = "Price vs Income by Fishing Mode", x = "Price", y = "Income") +
  theme_minimal()
```

```r
colSums(is.na(data))
```

```r
cleaned_data <- data %>%
  filter(price < quantile(price, 0.99) & catch < quantile(catch, 0.99))
cleaned_data
```

```r
summary(cleaned_data)
```

```r
numeric_vars <- select(data, price, catch, pbeach, ppier, pboat, pcharter, cbeach, cpier, cboat, ccharter, income)
numeric_vars
cor_matrix <- cor(numeric_vars)
cor_matrix
```

```r
print(cor_matrix)
```
